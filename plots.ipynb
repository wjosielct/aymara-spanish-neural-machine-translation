```python
# Librerías a usar
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Transformer (Arquitectura Estándar)

* Dimensión de representación interna ($d_{\text{model}}$): 512.
* Número de capas ($N$): 6 capas en el codificador / 6 capas en el decodificador.
* Número de cabezas en MHA ($h$): 8 cabezas de atención.
* Dimensión de red feed-forward ($d_{ffn}$): 2048.
* Dropout: 0.1.

## Aymara ${\rightarrow}$ Español

```python
results = pd.read_csv("results/metrics_AymaraToSpanish_Standard.csv")
results.head(10)
```

### val_loss vs epochs

```python
plt.figure()
plt.plot(results["epoch"], results["val_loss"])
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.title("Val Loss (Aymara → Español) - Estandard")
plt.show()
```

### val_ppl vs epochs

```python
plt.figure()
plt.plot(results["epoch"], results["val_ppl"])
plt.xlabel("Epoch")
plt.ylabel("Validation Perplexity (PPL)")
plt.title("Val PPL (Aymara → Español) - Estandard")
plt.show()
```

### val_bleu vs epochs

```python
bleu_df = results.dropna(subset=["val_bleu"])

plt.figure()
plt.plot(bleu_df["epoch"], bleu_df["val_bleu"], marker="o")

plt.xlabel("Epoch")
plt.ylabel("Validation BLEU")
plt.title("Val BLEU (Aymara → Español) - Estandard")

# 👇 Forzar ticks cada 10 épocas
plt.xticks(np.arange(0, bleu_df["epoch"].max() + 1, 10))

plt.show()
```
